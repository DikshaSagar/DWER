## Data Access via THREDDS / OpenDAP

This notebook demonstrates how to access climate model data hosted on the
NCI THREDDS server using OpenDAP. Data are accessed remotely using `xarray`,
allowing inspection and analysis without downloading full NetCDF files
locally.

The approach shown here is intentionally generic: dataset URLs, variables,
and time ranges can be adjusted to work with different models, experiments,
or future datasets hosted on the same infrastructure.
