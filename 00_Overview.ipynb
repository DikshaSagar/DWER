# Western Australia Climate Data Analysis (Prototype)

## Overview

This repository provides a set of reusable Jupyter notebooks demonstrating
how climate model data hosted on the NCI THREDDS (OpenDAP) infrastructure
can be accessed, analysed, and visualised using Python.

The notebooks are designed as a lightweight, extensible analysis framework
for rapid exploratory analysis and internal workflows, with minimal
assumptions about the underlying dataset beyond standard CF-compliant
metadata.

## Design Intent

The primary design goals of this work are to:

- Enable direct access to remotely hosted climate datasets without local downloads
- Provide clear, readable Jupyter notebook examples suitable for technical users
- Support portable execution via GitHub/Binder for easy sharing and reuse
- Ensure analysis logic is generic and easily transferable to future datasets

While the current examples use NARCliM 2.0 outputs, the structure and methods
are intentionally dataset-agnostic and are expected to be directly applicable
to future climate datasets hosted on NCI, including CSI products.

## Scope

The initial focus is on Western Australia, with notebooks organised to reflect
common climate analysis groupings such as mean climate conditions, temperature
extremes, precipitation extremes, and regional summary statistics.
