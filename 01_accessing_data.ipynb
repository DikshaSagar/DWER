# Task 2 â€“ Accessing NARCliM 2.0 Data via NCI THREDDS (OPeNDAP)

This notebook demonstrates how to access NARCliM 2.0 climate data hosted on the NCI THREDDS server using OPeNDAP and **xarray**.

The purpose of this notebook is to provide a **simple, reusable base example** that stakeholders and future users can adapt for analysis, without downloading large NetCDF files locally.

```python
from climate_utils import kelvin_to_celsius
import xarray as xr
import matplotlib.pyplot as plt
```

## Define the THREDDS OPeNDAP base URL

This is the stable base path for NARCliM 2.0 CMIP6 downscaled data on NCI. All datasets are accessed by appending directory components and a filename to this base URL.

```python
BASE_URL = (
    "https://dapds00.nci.org.au/thredds/dodsC/zz63/"
    "NARCliM2-0/output-CMIP6/DD"
)

BASE_URL
```

## Define dataset parameters

The following parameters describe the dataset location within the THREDDS directory structure. These are treated as directory labels rather than scientific inputs.

```python
domain = "AUS-18"
organisation = "NSW-Government"

gcm = "ACCESS-ESM1-5"
rcm = "NARCliM2-0-WRF412R3"
experiment = "ssp126"
variant = "r6i1p1f1"

frequency = "mon"
variable = "tasmax"
version = "v1-r1"
```

## Construct the OPeNDAP dataset URL

Using the base URL and parameters above, we construct the full file-level OPeNDAP endpoint required by xarray.

```python
file_url = (
    f"{BASE_URL}/"
    f"{domain}/{organisation}/"
    f"{gcm}/{experiment}/{variant}/"
    f"{rcm}/{version}/{frequency}/{variable}/latest/"
    f"{variable}_{domain}_{gcm}_{experiment}_{variant}_"
    f"{organisation}_{rcm}_{version}_{frequency}_201501-201512.nc"
)

file_url
```

## Open the dataset via OPeNDAP

The dataset is accessed directly from NCI without downloading the NetCDF file.

```python
ds = xr.open_dataset(file_url)
ds
```

## Simple direct-access example

For comparison, the same dataset can also be accessed using a fully hard-coded OPeNDAP URL. This represents the simplest possible access method.

```python
direct_url = (
    "https://dapds00.nci.org.au/thredds/dodsC/zz63/"
    "NARCliM2-0/output-CMIP6/DD/AUS-18/NSW-Government/"
    "ACCESS-ESM1-5/ssp126/r6i1p1f1/"
    "NARCliM2-0-WRF412R3/v1-r1/mon/tasmax/latest/"
    "tasmax_AUS-18_ACCESS-ESM1-5_ssp126_r6i1p1f1_"
    "NSW-Government_NARCliM2-0-WRF412R3_v1-r1_mon_201501-201512.nc"
)

ds_simple = xr.open_dataset(direct_url)
ds_simple
```

## Temperature unit standardisation

For consistency and accessibility to non-technical users, all temperature
variables are standardised to degrees Celsius (Â°C).

If the native dataset is provided in Kelvin (K), values are converted
using the standard relationship:

Â°C = K âˆ’ 273.15

The conversion is applied once using a shared utility function to avoid
duplication and reduce the risk of inconsistent handling across notebooks.


```python
ds["tasmax"] = kelvin_to_celsius(ds["tasmax"])

```

```python
# -------------------------------------------
# Dataset inspection & metadata sanity check
# -------------------------------------------

var = ds[variable]

print("Variable:", variable)
print("Dimensions:", var.dims)
print("Coordinates:", list(var.coords))
print("Units:", var.attrs.get("units"))
print("Long name:", var.attrs.get("long_name"))

# Determine available time range directly from dataset
start_time = var.time.min().values
end_time = var.time.max().values

print("\nTime coverage:")
print(start_time, "to", end_time)

# Use dataset-native time range for sanity stats
subset_preview = var.sel(time=slice(start_time, end_time))

print("\nSanity statistics:")
print("  Min:", float(subset_preview.min()))
print("  Max:", float(subset_preview.max()))
print("  Mean:", float(subset_preview.mean()))
print("  Missing values:", int(subset_preview.isnull().sum()))
```
